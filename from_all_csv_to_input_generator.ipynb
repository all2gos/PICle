```python
from IPython.display import HTML
import random

display(HTML("<style>.container { width:80% !important; }</style>"))

def hide(for_next=False, t = 'Toggle show/hide'):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = t
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ''
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)

hide(t='setup and import stuff')
```

## Średnia odległość

nowa cecha do potencjalnego MLa -> średnia odległość atomu do którego przyłączona jest cząsteczka od atomów Ni/Cu
można mieć wrażenie, że można pominąć tutaj atom środkowy ale nic bardziej mylnego - bo raz będzie Cu a raz Ni
na początku funkcja zwracająca odległość między jednym dwoma podanymi atomami

```python
hide(for_next=True, t=' def distance(a,b); def mult24(n)')
```

```python
import numpy as np

def distance(a,b):   
    
    x_a, y_a, z_a = cluster_cor[a][0], cluster_cor[a][1], cluster_cor[a][2]
    x_b, y_b, z_b = cluster_cor[b][0], cluster_cor[b][1], cluster_cor[b][2]
    
    return np.sqrt((x_a - x_b)**2 + (y_a - y_b)**2 + (z_a - z_b)**2)

#sprawdziłem na kilku przykładach --> działa z jakąś tam dokładnością, ale chyba nic z tym nie mogę zrobić

#funkcja która przemnaża cordy *24 

def mult24(n):
    for i in range(len(n)):
        for j in range(3):
            n[i][j] = float(n[i][j])*24
    return n
```

```python
#na samym poczatku funckja ktora zamieni topologie, ktora przez eksport do csv zaczela byc stringiem spowrotem na liste

def topology_to_list(string):
    
    string_from_topology = []
    for i in range(2,80,6):
        string_from_topology.append(string[i:i+2])  
        
    return string_from_topology

#teraz funkcja, ktora zwroci dwie listy --> pozycje na ktorych znajduje się miedz i pozycje na ktorych znajduje sie nikiel

def position(cl):
    
    cu_pos = []
    ni_pos = []
        
    for atom in range(len(cl)):
        if cl[atom] == 'Ni':
            ni_pos.append(atom)
        else:
            cu_pos.append(atom)
            
    return cu_pos, ni_pos

#no i to co pozostało to funkcja zwracająca te średnie odległości

def avg_distance(cl):
    
    cl = topology_to_list(cl)
    cu_pos, ni_pos = position(cl)[0], position(cl)[1]
    
    
    avg_distance_list = []  
    
    #atom1 
    
    #nikiel
    
    atom_1_avg_distance = 1
    nb_of_nickel = 0
    
    print(ni_pos)
    print('dystanse atom1 od ni: ')
    
    for pos in ni_pos:        
        
        
        if pos != 1:
            atom_1_avg_distance *= np.sqrt(distance(1,pos))
            nb_of_nickel += 1
            print('1,' + str(pos), end ='  ')
    print()  
    avg_distance_list.append(atom_1_avg_distance**(1/nb_of_nickel))
    
    
    
    #pozostale srednie odleglosci liczy sie analogicznie z drobnymi poprawkami
    
    #miedz
    
    atom_1_avg_distance = 1
    nb_of_copper = 0
    
    print()
    print(cu_pos)
    print('dystanse atom 1 od miedzi: ')
    
    for pos in cu_pos:        
        
        
        if pos != 1:
            atom_1_avg_distance *= np.sqrt(distance(1,pos))
            nb_of_copper += 1
            print('1,' + str(pos), end ='  ')
    print('')
    print(atom_1_avg_distance)
    avg_distance_list.append(atom_1_avg_distance**(1/nb_of_copper))
    
    #atom2
    
    #nickel 
    
    atom_2_avg_distance = 1
    nb_of_nickel = 0  
    
    print()
    print(ni_pos)
    print('dystanse atom 2 od niklu: ')
    
    for pos in ni_pos:         
        
        if pos != 2:
            atom_2_avg_distance *= np.sqrt(distance(2,pos))
            nb_of_nickel += 1   
            print('2,' + str(pos), end ='  ')
    print('')
    avg_distance_list.append(atom_2_avg_distance**(1/nb_of_nickel))
    
    #miedz
    
    atom_2_avg_distance = 1
    nb_of_copper = 0 
    
    print()
    print(cu_pos)
    print('dystanse atom 2 od miedzi: ')
    
    for pos in cu_pos:          
        if pos != 2:
            atom_2_avg_distance *= np.sqrt(distance(2,pos))
            nb_of_copper += 1
            print('2,' + str(pos), end ='  ')
    print('')
    print(atom_2_avg_distance)
        
    avg_distance_list.append(atom_2_avg_distance**(1/nb_of_copper))
    
    return avg_distance_list
        
```

```python
#zmiana funkcji distance - przyjmuje teraz 3 argument - listę kordów atomowych

def distance(a,b,cluster_cor):   
    
    x_a, y_a, z_a = cluster_cor[a][0], cluster_cor[a][1], cluster_cor[a][2]
    x_b, y_b, z_b = cluster_cor[b][0], cluster_cor[b][1], cluster_cor[b][2]
    
    return np.sqrt((x_a - x_b)**2 + (y_a - y_b)**2 + (z_a - z_b)**2)

#myślę, że sensowne będzie sprawdzanie odległości każdego pierwiastka C/O/H od atomu 1 i 2 w ten sposob sprawdze
#jak bardzo usrednianie tej odleglosci zaniedbuje rzeczywistosc

def cmplx_identity(cluster,num_cluster):
    
    cmplx_i = []
    
    top = topology_to_list(df['topologia'][num_cluster])
    nb_nickel = 13-df['liczba Ni'][num_cluster]
    
    #jest to konieczny zabieg, nie moge sobie przyjac 1 i 2 pozycji poniewaz zmieniam te pozycje w locie xDDDD
    #to co tutaj się dzieje to ustalam, który atom jest pierwszy, drugi i środkowy
    #jak wiadomo w VASPie generując INPUTy mocno żongluje tymi coordami
    #wiec musze w tym momencie ustalic na podstawie jakich cordow mam wyznaczac odleglosci
    
    if top[0] == 'Cu' and top[1] == 'Cu' and top[2] =='Cu':
        first_atom, second_atom, centre_atom = 1,2,0
        
    elif top[0] == 'Cu' and top[1] == 'Cu' and top[2] =='Ni':
        first_atom, second_atom, centre_atom = 1, nb_nickel, 0
        
    elif top[0] == 'Cu' and top[1] == 'Ni' and top[2] =='Cu':
        first_atom, second_atom, centre_atom = 1, nb_nickel,0
        
        #wydaje sie git
    elif top[0] == 'Cu' and top[1] == 'Ni' and top[2] =='Ni':
        first_atom, second_atom, centre_atom = nb_nickel, nb_nickel+1, 0
        
        #wydaje sie git
    elif top[0] == 'Ni' and top[1] == 'Cu' and top[2] =='Cu':
        first_atom, second_atom, centre_atom = 0,1,nb_nickel
        
    elif top[0] == 'Ni' and top[1] == 'Cu' and top[2] =='Ni':
        first_atom, second_atom, centre_atom = 0, nb_nickel
        
    elif top[0] == 'Ni' and top[1] == 'Ni' and top[2] =='Cu':
        first_atom, second_atom, centre_atom = 0, nb_nickel+1, centre_atom
        
    elif top[0] == 'Ni' and top[1] == 'Ni' and top[2] =='Ni':
        first_atom, second_atom, centre_atom = nb_nickel+1, nb_nickel+2, nb_nickel
        
    
    #print(first_atom)
    #print(second_atom)
    
    #odległości od 1 atomu
    cmplx_i.append(f'{distance(first_atom,-1,cluster):.4}')
    cmplx_i.append(f'{distance(first_atom,-2,cluster):.4}')
    cmplx_i.append(f'{distance(first_atom,-3,cluster):.4}')
    
    if len(cluster) == 17:
        cmplx_i.append(f'{distance(first_atom,-4,cluster):.4}')
    
    #odległości od 2 atomu
    cmplx_i.append(f'{distance(second_atom,-1,cluster):.4}')
    cmplx_i.append(f'{distance(second_atom,-2,cluster):.4}')
    cmplx_i.append(f'{distance(second_atom,-3,cluster):.4}')
    
    if len(cluster) == 17:
        cmplx_i.append(f'{distance(second_atom,-4,cluster):.4}')
    
    #odleglosci od srodkowego atomu
    
    cmplx_i.append(f'{distance(centre_atom,-1,cluster):.4}')
    cmplx_i.append(f'{distance(centre_atom,-2,cluster):.4}')
    cmplx_i.append(f'{distance(centre_atom,-3,cluster):.4}')
    
    if len(cluster) == 17:
        cmplx_i.append(f'{distance(centre_atom,-4,cluster):.4}')
        
    #print(cmplx_i)
    return cmplx_i
```

```python
def generowanie_srednich_odleglosci(klaster, typ_polaczenia, nr_klastra):
    
    #pobieram infromacje o liczbie atomów niklu i miedzi w klastrze
    ni = df['liczba Ni'][nr_klastra]
    cu = 13 - ni
    
    #jeżeli poniższy typ połączenia jest poprawny    
    if typ_polaczenia == 'MOCM':
        
        # tworzę dwie listy zawierające informacje o odległościach atomów przyłączanej
        # cząsteczki w czystym klastrze miedziowym i niklowym
        l1 = cmplx_identity(clear_Cu_MOCM_opt,0)
        l2 = cmplx_identity(clear_Ni_MOCM_opt,191)
        l_avg = []
        
        #tworzę listę która jest średnią ważoną tych dwóch poprzednich list 
        for i in range(len(l1)):
            l_avg.append(f'{((ni*float(l1[i])+cu*float(l2[i]))/13):.4}')
            
    elif typ_polaczenia == 'MOCOM':
        l1 = cmplx_identity(clear_Cu_MOCOM_opt,0)
        l2 = cmplx_identity(clear_Ni_MOCOM_opt,191)
        l_avg = []
        for i in range(len(l1)):
            l_avg.append(f'{((ni*float(l1[i])+cu*float(l2[i]))/13):.4}')
            
    elif typ_polaczenia == 'MOC_':
        l1 = cmplx_identity(clear_Cu_MOC__opt,0)
        l2 = cmplx_identity(clear_Ni_MOC__opt,191)
        l_avg = []
        for i in range(len(l1)):
            l_avg.append(f'{((ni*float(l1[i])+cu*float(l2[i]))/13):.4}')
        
    elif typ_polaczenia == 'MCM':
        l1 = cmplx_identity(clear_Cu_MCM_opt,0)
        l2 = cmplx_identity(clear_Ni_MCM_opt,191)
        l_avg = []
        for i in range(len(l1)):
            l_avg.append(f'{((ni*float(l1[i])+cu*float(l2[i]))/13):.4}')
        
    elif typ_polaczenia == 'MO':
        l1 = cmplx_identity(clear_Cu_MO_opt,0)
        l2 = cmplx_identity(clear_Ni_MO_opt,191)
        l_avg = []
        for i in range(len(l1)):
            l_avg.append(f'{((ni*float(l1[i])+cu*float(l2[i]))/13):.4}')
        
    elif typ_polaczenia == 'MOCM_H_n_bond':
        l1 = cmplx_identity(clear_Cu_MOCM_H_n_bond_opt,0)
        l2 = cmplx_identity(clear_Ni_MOCM_H_n_bond_opt,191)
        l_avg = []
        for i in range(len(l1)):
            l_avg.append(f'{((ni*float(l1[i])+cu*float(l2[i]))/13):.4}')
        
    elif typ_polaczenia == 'MOCM_H_bond':
        l1 = cmplx_identity(clear_Cu_MOCM_H_bond_opt,0)
        l2 = cmplx_identity(clear_Ni_MOCM_H_bond_opt,191)
        l_avg = []
        for i in range(len(l1)):
            l_avg.append(f'{((ni*float(l1[i])+cu*float(l2[i]))/13):.4}')
        
    elif typ_polaczenia == 'MOC':
        l1 = cmplx_identity(clear_Cu_MOC_opt,0)
        l2 = cmplx_identity(clear_Ni_MOC_opt,191)
        l_avg = []
        for i in range(len(l1)):
            l_avg.append(f'{((ni*float(l1[i])+cu*float(l2[i]))/13):.4}')
        
    elif typ_polaczenia == 'MC':
        l1 = cmplx_identity(clear_Cu_MC_opt,0)
        l2 = cmplx_identity(clear_Ni_MC_opt,191)
        l_avg = []
        for i in range(len(l1)):
            l_avg.append(f'{((ni*float(l1[i])+cu*float(l2[i]))/13):.4}')
    
    #lista pomocniczna
    compares = []
    
    #tworzę listę, która zawiera informacje o średnich odległośiach między atomami
    #klastra a badanym indywiduum
    opt_ls = cmplx_identity(klaster,nr_klastra)
    
    #do listy pomocnicznej dodaję informację o różnicy między poszczególnymi odległościami
    #w przypadku klastra optymalizowanego i odległości wyznaczonych na podstawie
    #czystych klastrów 
    
    for i in range(len(opt_ls)):
        
        #jeżeli chcemy bez modułu to zostawiamy tą linię
#         compares.append(float(f'{float(opt_ls[i])-float(l_avg[i]):.3}'))
        
        #jeżeli chcemy z modułem to zostawiamy tą
        
        compares.append(np.sqrt((float(f'{float(opt_ls[i])-float(l_avg[i]):.3}'))**2))
    name = str(nr_klastra)+'_'+ typ_polaczenia
    return compares, len(compares), name

#funkcja, która będzie printować staty jednego przypadku
```

```python
def avogadro(ls, Ni):
    ls = str(ls)
    avogadro_ls = ''
    for char in ls:
        if char not in ['[',']',',',"'"]:
            avogadro_ls += char
    avogadro_ls = avogadro_ls.split(' ')  
    
    atom_ls = []
    
    licznik = 0
    for cor in avogadro_ls:      
            
        if len(atom_ls) !=3:
            atom_ls.append(cor)
        else:  
            
            if licznik <40-3*Ni:
                atom_ls.insert(0,'Cu')
            elif licznik <40:
                atom_ls.insert(0,'Ni')            
            elif licznik <43:
                atom_ls.insert(0,'C')
            elif licznik <49:
                atom_ls.insert(0,'O')
            else:
                atom_ls.insert(0,'H')
            print('    '.join(atom_ls))            
            atom_ls = []
            atom_ls.append(cor)
        licznik +=1  
    #ostatni atom
    if licznik >50:
        atom_ls.insert(0,'H')
    else:
        atom_ls.insert(0,'O')
    print('    '.join(atom_ls))      
```
